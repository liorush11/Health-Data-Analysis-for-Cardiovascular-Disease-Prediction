1. ניתוח ערכים חריגים ב-BMI:


```python

import seaborn as sns
import matplotlib.pyplot as plt


# גרף התפלגות BMI
plt.figure(figsize=(8, 6))
sns.histplot(data=data_df, x='BMI', kde=True, bins=30)
plt.title('Distribution of BMI')
plt.xlabel('BMI')
plt.ylabel('Frequency')
plt.show()


# זיהוי ערכים חריגים
q1_bmi = data_df['BMI'].quantile(0.25)
q3_bmi = data_df['BMI'].quantile(0.75)
iqr_bmi = q3_bmi - q1_bmi


lower_bound_bmi = q1_bmi - 1.5 * iqr_bmi
upper_bound_bmi = q3_bmi + 1.5 * iqr_bmi


outliers_bmi = data_df[(data_df['BMI'] < lower_bound_bmi) | (data_df['BMI'] > upper_bound_bmi)]
print(f"Number of outliers in BMI: {len(outliers_bmi)}")


# יצירת גרף Boxplot עבור BMI
plt.figure(figsize=(8, 6))
sns.boxplot(data=data_df, x='BMI')
plt.title('Boxplot of BMI')
plt.xlabel('BMI')
plt.show()
```

3. ניתוח לחץ דיאסטולי:


```python
# חישוב ערכים סטטיסטיים
mean_bp_dias = data_df['bp_dias'].mean()  # ממוצע של לחץ דם דיאסטולי
median_bp_dias = data_df['bp_dias'].median()  # חציון של לחץ דם דיאסטולי
std_bp_dias = data_df['bp_dias'].std()  # סטיית תקן של לחץ דם דיאסטולי
min_bp_dias = data_df['bp_dias'].min()  # ערך מינימלי
max_bp_dias = data_df['bp_dias'].max()  # ערך מקסימלי
q1_bp_dias = data_df['bp_dias'].quantile(0.25)  # רבעון ראשון
q3_bp_dias = data_df['bp_dias'].quantile(0.75)  # רבעון שלישי
iqr_bp_dias = q3_bp_dias - q1_bp_dias  # טווח בין-רבעוני (IQR)


# הדפסת הערכים הסטטיסטיים
print(f"Mean Diastolic BP: {mean_bp_dias}")
print(f"Median Diastolic BP: {median_bp_dias}")
print(f"Standard Deviation Diastolic BP: {std_bp_dias}")
print(f"Min Diastolic BP: {min_bp_dias}")
print(f"Max Diastolic BP: {max_bp_dias}")
print(f"IQR Diastolic BP: {iqr_bp_dias}")


# זיהוי ערכים חריגים
lower_bound_bp_dias = q1_bp_dias - 1.5 * iqr_bp_dias  # גבול תחתון לערכים חריגים
upper_bound_bp_dias = q3_bp_dias + 1.5 * iqr_bp_dias  # גבול עליון לערכים חריגים
outliers_bp_dias = data_df[(data_df['bp_dias'] < lower_bound_bp_dias) | (data_df['bp_dias'] > upper_bound_bp_dias)]  
# סינון הערכים החריגים על בסיס גבולות


# הדפסת מספר הערכים החריגים
print(f"Number of outliers in Diastolic BP: {len(outliers_bp_dias)}")


# גרף התפלגות
plt.figure(figsize=(8, 6))
sns.histplot(data=data_df, x='bp_dias', kde=True, bins=30)  # היסטוגרמה עם קו צפיפות
plt.title('Distribution of Diastolic BP')  # כותרת הגרף
plt.xlabel('Diastolic BP')  # תווית לציר ה-X
plt.ylabel('Frequency')  # תווית לציר ה-Y
plt.show()


# גרף Boxplot
plt.figure(figsize=(8, 6))
sns.boxplot(data=data_df, x='bp_dias')  # Boxplot לערכי לחץ דם דיאסטולי
plt.title('Boxplot of Diastolic BP')  # כותרת הגרף
plt.xlabel('Diastolic BP')  # תווית לציר ה-X
plt.show()
```

```python
# שמירת רק העמודות הנומריות
numeric_df = data_df.select_dtypes(include=['float64', 'int64'])


# חישוב מטריצת הקורלציות
correlation_matrix = numeric_df.corr()


# מציאת כל הזוגות עם קורלציה מעל 0.7 או מתחת ל- -0.7 (ללא הקורלציה העצמית של 1)
high_corr_pairs = correlation_matrix.unstack().reset_index()
high_corr_pairs.columns = ['Variable 1', 'Variable 2', 'Correlation']
high_corr_pairs = high_corr_pairs[(high_corr_pairs['Correlation'] > 0.7) | (high_corr_pairs['Correlation'] < -0.7)]
high_corr_pairs = high_corr_pairs[high_corr_pairs['Variable 1'] != high_corr_pairs['Variable 2']]  # הסרת קורלציה עצמית
high_corr_pairs = high_corr_pairs.drop_duplicates(subset=['Correlation'])  # מניעת כפילויות


# יצירת Heatmap נפרד לכל זוג משתנים עם קורלציה גבוהה
for index, row in high_corr_pairs.iterrows():
    var1 = row['Variable 1']
    var2 = row['Variable 2']
    corr_value = row['Correlation']
   
    # חישוב מטריצה עבור המשתנים
    mini_corr_matrix = correlation_matrix.loc[[var1, var2], [var1, var2]]
   
    # יצירת Heatmap
    plt.figure(figsize=(6, 4))
    sns.heatmap(mini_corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1, cbar=True, square=True)
    plt.title(f'Heatmap of Correlation: {var1} vs {var2} (Corr={corr_value:.2f})')
    plt.show()
```
